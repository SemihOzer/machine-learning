Tile Dataset

```python
Xt = np.tile(X,(10,1))
Yt= np.tile(Y,(10,1))   
```

Create model

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.Input(shape=(2,)))
model.add(tf.keras.layers.Dense(32, activation='relu',name="l1"))
model.add(tf.keras.layers.Dense(25,activation="relu",name="l2"))
model.add(tf.keras.layers.Dense(1,activation="sigmoid",name="l3"))


```

Compile and fit model

```python
model.compile(
    loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
)

model.fit(
    Xt,Yt,            
    epochs=10,
    steps_per_epoch=200
    
)
```

Predictions

```python
X_test = np.array([
    [200,13.9],  
    [200,17]])   
predictions = model.predict(X_test)

result = np.zeros_like(predictions)

for i in range(len(predictions)):
  if predictions[i] >= 0.5:
    result[i] = 1
  else:
    result[i] = 0
  
print(result)
```
